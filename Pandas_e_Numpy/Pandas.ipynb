## VariÃ¡veis Dummy | HotEncoder

Para o programa de machine learning que usaremos no futuro quando formos analisar os itens da coluna sexo por exemplo, devemos passar para zeros e ums.
Por que? Porque matematicamente o programa nao sabe que mulher Ã© mulher e que homem Ã© homem por serem strings, para isso devemos dizer se Ã© 1(True) ou 0(False).
Dei o exemplo da coluna 'Sexo', mas serve para outras tambem

##### Como podemos ver a baixo, transformamos a coluna sexo em 2 colunas, uma para homem e outra para mulher. Vai ser verdadeiro quem tiver o 1(true), mas o ideal Ã© trasformar em apenas 1 coluna com zeros e ums, porque nao faz sentido ter 2 colunas que matematicamente me mostram o mesmo resultado, para isso fazer usar o comando 'drop_first=True' que 'dropa' a primeira coluna e deixa apenas a segunda

```python
pd.get_dummies(data, columns=['Sexo'])
```

```python
data = pd.get_dummies(data, columns=['Sexo'], drop_first=True)  # Estou colocando o data para armazenar no dataframe
```

```python
data
```

```python
pd.get_dummies(data['Embarque'], drop_first=True)  # Aqui ele tira a coluna 'C' com o dropfirst, se nao Ã© 'Q' e nem 'S', Ã© 'C'
```

```python
embarque = pd.get_dummies(data['Embarque'], drop_first=True)  # Colocando em uma variavel para concatenar depois
```

##### Dessa maneira que usamos, fazemos isso para extrair somente o titulo das pessoas (miss, mister, master e etec)
Acima temos uma breve lista das expressoes regulares, mas vou explicar somente essas que usamos a baixo.
'a-zA-Z' - aqui estamos pegando tudo entre a e z maiusculo e minusculo
[] - caracteres em colchete
() - grupo
+ - para pegar mais de 1 caracter
. - com o ponto eu digo que quero ele como referencia, entao tudo antes do ponto
\. - por estar justamente usando um ponto, devemos usar a barra invertida para o interpretador nao confudir com a expressao regular '.' que significa 'qualquer caractere'

```python
data['Titulo'] = data.Nome.str.extract('([a-zA-Z]+)\.')
data.Titulo
```

## Mudando Valores Linha Por Linha - iterrows()

So mais uma forma de mudar valores. Vou antes disso criar um novo dataframe que vai se chamar 'cabine' para voltarmos para o modo original do arquivo

```python
cabine = pd.read_csv('titanic/train.csv')
```

```python
cabine.Cabin
```

```python
for idx, _ in cabine[['Cabin']].dropna().iterrows():  # O _ Ã© porque nao estamos passando nada | O dropna exclui todas as linhas que sao nulos
    cabine_individual = cabine['Cabin'].at[idx]
    cabine['Cabin'].at[idx] = cabine_individual[0]
```

```python
cabine[['Cabin']].dropna()  # Com o cod acima pegamos apenas a primeira letra dos itens da 'Cabin'
```

```python
cabine[['Cabin', 'Survived', 'Sex', 'Pclass']].dropna().groupby(['Cabin', 'Sex']).mean()
```

Acima podemos qual o percentual de pessoas que morreram e sobreviveram por cabine
